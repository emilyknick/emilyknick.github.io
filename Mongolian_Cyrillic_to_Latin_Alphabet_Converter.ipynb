## How to use the converter

1.   First, run the code cell labeled **Set-up** \(directly below\) to import necessary modules/packages and set-up the mappings between the Cyrillic and Latin alphabets.

2.   Next, if you want to convert **copy-pasted text** from Mongolian Cyrillic to the Latin Alphabet:
* Run the code cell labeled **Basic Text Converter**.
* Enter your text in the input box that appears below, and click the "Convert" button.

3. Instead, if you want to convert **uploaded .txt files** from Mongolian Cyrillic to the Latin Alphabet:
* Upload your .txt files to the session storage by clicking on the file icon in the righthand sidebar.
* Run the code cell labeled **.txt File Converter**.
* Download the output files by clicking on the file icon in the righthand sidebar.





```python
##################
##### Set-up #####
##################

import ipywidgets as widgets
from IPython.display import display, HTML, Markdown
from ipywidgets import interact, FloatText, Button, VBox, HBox, Output, Label
import glob
from pathlib import Path

mongolianCyrillic_latin = {
    "А": "A", "а": "a",
    "Б": "B", "б": "b",
    "В": "V", "в": "v",
    "Г": "G", "г": "g",
    "Д": "D", "д": "d",
    "Е": "Ye", "е": "ye",
    "Ё": "Yo", "ё": "yo",
    "Ж": "J", "ж": "j",
    "З": "Z", "з": "z",
    "И": "I", "и": "i",
    "Й": "I", "й": "i",
    "К": "K", "к": "k",
    "Л": "L", "л": "l",
    "М": "M", "м": "m",
    "Н": "N", "н": "n",
    "О": "O", "о": "o",
    "Ө": "Ö", "ө": "ö",
    "П": "P", "п": "p",
    "Р": "R", "р": "r",
    "С": "S", "с": "s",
    "Т": "T", "т": "t",
    "У": "U", "у": "u",
    "Ү": "Ü", "ү": "ü",
    "Ф": "F", "ф": "f",
    "Х": "Kh", "х": "kh",
    "Ц": "Ts", "ц": "ts",
    "Ч": "Ch", "ч": "ch",
    "Ш": "Sh", "ш": "sh",
    "Щ": "Shch", "щ": "shch",
    "Ъ": "",  "ъ": "",
    "Ы": "Ii", "ы": "ii",
    "Ь": "",  "ь": "",
    "Э": "E", "э": "e",
    "Ю": "Yu", "ю": "yu",
    "Я": "Ya", "я": "ya"
}
```

```python
################################
##### Basic Text Converter #####
################################

#Interactive converter display.

converter_description = widgets.Label(value="Enter the Mongolian text you'd like to convert from Cyrillic to the Latin alphabet:")

#Create input widget.

mongolianCyrillic_input = widgets.Textarea(
    value = "",
    style={'description_width': '150px'},
    layout={'width': '750px', 'height': '100px'}
)

convert_button = Button(description="Convert", button_style='info')
latinText_display = Output()

#Define a version of mongolian_to_latin() that works with ipywidgets.

def mongolian_to_latinWidget(button):
  with latinText_display:
    latinText_display.clear_output()

    output = []
    text = mongolianCyrillic_input.value
    for letter in text:
        output.append(mongolianCyrillic_latin.get(letter, letter))
    print("".join(output))

convert_button.on_click(mongolian_to_latinWidget)
display(VBox([converter_description, mongolianCyrillic_input, convert_button, latinText_display]))
print(mongolianCyrillic_input.value)
```

```python
###############################
##### .txt File Converter #####
###############################

def mongolian_to_latin():
    files = []
    files = glob.glob('*.txt')

    print(f"{len(files)} file(s) uploaded for conversion.")

    for txt in files:
      txtString = Path(txt).read_text()
      latin_output = []
      for letter in txtString:
        latin_output.append(mongolianCyrillic_latin.get(letter, letter))
      latin_output = "".join(latin_output)
      alt_fileName = txt.replace(".txt", "")
      output_fileName = f"{alt_fileName}_latinAlphabet.txt"
      with open(output_fileName, 'w') as file:
        file.write(latin_output)
    print(f"File(s) converted.")

mongolian_to_latin()
```
